##### Correlação de Colunas - Preço de Casas na Índia

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data/House Price India.csv")
```

##### Correlações

```python
corr_matrix = df.corr()
corr_matrix
```

```python
plt.style.use('dark_background')
```

```python
plt.figure(figsize=(24, 10))
heatmap = sns.heatmap(corr_matrix, annot=True, linewidths=.5, fmt='.2f')
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0, ha='right')
plt.show()
```

```python
corr = df.corrwith(df['Price']).sort_values(ascending=False).to_frame()
```

```python
plt.figure(figsize=(8, 10))
heatmap = sns.heatmap(corr, annot=True, linewidths=.5, fmt='.2f')
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0, ha='right')
plt.show()
```
